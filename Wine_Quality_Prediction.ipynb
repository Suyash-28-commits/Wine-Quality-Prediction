```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('/content/drive/MyDrive/WineQT.csv')
```

```python
wine.sample(5)
```

```python
#Standerdization (Z-Score Normalization)
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(wine.drop('quality',axis=1),wine['quality'],test_size=0.2,random_state=42)
X_train.shape , X_test.shape
```

```python
#Standard Scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print("Score on actual data",accuracy_score(y_test,y_pred))
print("Score on scaled data",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeRegressor
dr = DecisionTreeRegressor()
dr_scaled = DecisionTreeRegressor()
```

```python
dr.fit(X_train,y_train)
dr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = dr.predict(X_test)
y_pred_scaled = dr_scaled.predict(X_test_scaled)
```

```python
print("Actual score: ",accuracy_score(y_test,y_pred))
print("Scaled score: ",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn_scaled = KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
knn_scaled.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred_scaled = knn.predict(X_test_scaled)
```

```python
print("Actual Score: ",accuracy_score(y_test,y_pred))
print("Scaled Score: ",accuracy_score(y_test,y_pred_scaled))
```
